## Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial_excluded.csv', sep=';')
dados.head(10)
```

#### https://pandas.pydata.org/pandas-docs/stable/reference/index.html 

```python
dados['Valor'].mean()
```

```python
bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
```

```python
dados['Bairro'].drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro')
```

```python
type(grupo_bairro)
```

```python
grupo_bairro.groups
```

```python
for bairro, data in grupo_bairro:
    print('{} -> {}'.format(bairro, data.Valor.mean()))
```

```python
grupo_bairro['Valor'].mean().round(2)
```

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2)
```

## Estatística Descritiva

```python
grupo_bairro['Valor'].describe().round()
```

```python
grupo_bairro['Valor'].aggregate(['min','max']).rename(columns = {'min':'Minimo','max':'Maximo'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20, 10))
```

```python
fig = grupo_bairro['Valor'].std().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize':22})
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize':22})
```

```python
fig = grupo_bairro['Valor'].max().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel')
fig.set_title('Valor Maximo de Aluguel por Bairro', {'fontsize':22})
```
