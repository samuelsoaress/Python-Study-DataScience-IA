# Criando Novas Variaveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('aluguel_residencial_snulls.csv',sep = ';')
```

```python
dados.head(10)
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor'] / dados['Area']
```

```python
dados.head(10)
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2) # round arredonda em 2 casas decimais
```

```python
dados.head(10)
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(2)
```

```python
dados.head(10)
```

```python
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x:'Casa' if x in casa else 'Apartamento') # Função apply vai aplicar a função em cada linha que ele tiver
# nesse caso ele vai mudar pra casa se no valor de tipo estiver algum valor da lista casa
```

```python
dados.head(20)
```
